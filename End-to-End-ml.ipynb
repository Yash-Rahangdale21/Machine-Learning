```python
import numpy as np 
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
#OverView of data set
df.head()
```

```python
## for check missing values 
df.info()
```

```python
# preprocess => useless remove is remove
df = df.iloc[:,1:] ## sare row chhiye one onward cols chhiye 
```

```python
df.head()
```

```python
#EDA 
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['cgpa'],df['iq'],c = df['placement'])
```

```python
# Extract input and outputs cols
## input are independent and output are dependent 
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
## Train test split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Scale the values
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
# Train the model  in that we decide we use classifier
from sklearn.linear_model import LogisticRegression
```

```python
#Steps
# 0 . preprocess + EDA + Feature Selection
# 1. Extract input and outputs cols
# 2. Scale the values 
# 3. Train test split
# 4. Train the model
# 5. Evaluate the model/model selection
# 6. Deploy the model
```

```python
clf = LogisticRegression()
```

```python
## this is model training 
clf.fit(X_train,y_train)
```

```python
# Evaluate the model/model selection accuracy nikalte test pr 
y_pred = clf.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, y_train.values, clf=clf, legend=2)
```
